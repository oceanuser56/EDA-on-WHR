```python
print(df.isnull().values.sum())
df=df.replace(" ",np.NaN)
df=df.fillna(df.mean())
print(df.isnull().values.sum())
```

**The first line prints the total number of missing values in the DataFrame. The second line replaces empty spaces with NaN (Not a Number) and**
**then fills the remaining NaN values with the mean of their respective columns. The final line prints the new total number of missing values after**
**the imputation.**

# Histogram

```python
import seaborn as sns
sns.distplot(df['fixed acidity'], hist=True, kde=True, bins=15, color='red', hist_kws={'edgecolor': 'black'})

# Adding labels
plt.title('acidity count')
plt.xlabel('score')
plt.ylabel('Count')

plt.show()
```

**The x-axis represents the acidity scores, while the y-axis shows the count of occurrences. The title of the plot is set as "Acidity Count," and** 
**the axes are appropriately labeled. This visualization aids in understanding the distribution pattern of fixed acidity in the dataset, including**
**the concentration of wines with specific acidity levels. The histogram reveals the frequency of different acidity scores, and the kernel density** **estimate provides a smooth curve, offering additional insights into the underlying distribution.**

# Bar Plot

```python
sns.barplot(data=df,x='volatile acidity',color='blue',edgecolor='black')
plt.title('volatile acidity count')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

**It helps in understanding the distribution and concentration of wines with varying volatile acidity levels.**

# Count Plot

```python
plt.figure(figsize=(15,15))
plt.title('Various wine quality', fontsize=20, fontweight='bold')
sns.countplot(y='quality', data=df, order=df['quality'].value_counts().index)
plt.xlabel('Count', fontsize=15, fontweight='bold')
plt.ylabel('quality', fontsize=20, fontweight='bold')
plt.show()
```

**Provides a visual representation of the distribution of wine quality levels in the dataset, helping to identify the prevalence of each quality category.**

# KDE Plot

```python
#seaborn kde
sns.kdeplot(data=df,x='residual sugar',color='red',shade='True',linewidth=2,edgecolor='black')
#add labels
plt.title('residual sugar')
plt.xlabel('distribution')
plt.ylabel('count')
```

**the resulting KDE plot visually represents the distribution of residual sugar values in the dataset. The shaded area under the curve provides an estimate of the probability density of observing different levels of residual sugar. The plot helps in understanding the shape and concentration of residual sugar values in the dataset.**

# Box Plot

```python
plt.figure(figsize=(5,5))
sns.boxplot(y='chlorides', data=df)
plt.show()
```

**It helps in visualizing the distribution and statistical characteristics of the 'chlorides' variable, providing insights into the presence of outliers and the overall variability of the data. Outliers in a box plot serve as important indicators of extreme values within a dataset.**

# Scatter Plot

```python
sns.scatterplot(x=df['density'],y=df['pH'])
plt.title('scatterplot of density vs pH')
plt.xlabel('density')
plt.ylabel('pH')
plt.show()
```

**It visualizes the relationship between these two variables based on the provided dataset. Each data point is represented as a point on the graph, where the x-axis corresponds to the density values, and the y-axis corresponds to the pH values.**

# Line Plot

```python
plt.figure(figsize=(15,15))
plt.title('Wine consumption', fontsize=20, fontweight='bold')
sns.lineplot(y='quality', x='alcohol', data=df)
plt.xlabel('alcohol', fontsize=15, fontweight='bold')
plt.ylabel('quality', fontsize=20, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

**The line plot effectively communicates how wine quality changes in relation to alcohol content, offering valuable insights into the correlation between these two variables within the dataset.**
