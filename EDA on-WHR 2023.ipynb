# HISTOGRAM

```python
import seaborn as sns
sns.distplot(df['Ladder score'], hist=True, kde=True, bins=15, color='red', hist_kws={'edgecolor': 'black'})

# Adding labels
plt.title('Ladder score count')
plt.xlabel('score')
plt.ylabel('Count')

plt.show()
```

**Ladder Score Distribution**

The code generates a distribution plot for the 'Ladder score,' a metric likely representing a country's subjective well-being or happiness. The plot combines a histogram and a kernel density estimate.

- **Histogram:** The red bars represent the frequency or count of countries falling within specific score intervals (bins). In this case, there are 15 bins, providing a visual representation of how the 'Ladder score' is distributed across the dataset.

- **Kernel Density Estimate (KDE):** The smooth red line (KDE) provides a continuous probability density function, giving insights into the likelihood of different 'Ladder score' vthe data.

**Conclusion:**
This visualization allows for a quick understanding of the distribution of 'Ladder scores' across countries in the dataset, providing insights into the general well-being or happiness levels reported.

# PIE-CHART

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (replace this with your dataset)
data = np.random.randn(1000)

# Create categories for the data
# For simplicity, we'll divide the data into two categories: positive and negative values
positive_data = data[data > 0]
negative_data = data[data <= 0]

# Calculate the sizes of the categories
positive_size = len(positive_data)
negative_size = len(negative_data)

# Create a pie chart using matplotlib
labels = ['Positive Values', 'Negative Values']
sizes = [positive_size, negative_size]
colors = ['blue', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Positive and Negative Values')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

**negative distribution in the pie chart is less (i.e. 49.4%) whereas the value of positive distribution is more(i.e. 50.6%). However, it's**
**important to note that the code is using randomly generated data (np.random.randn(1000)) rather than the specific dataset provided. Therefore,**
**the explanation will be general.**

# COUNT-PLOT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example:
df_copy = pd.DataFrame({'Result': ['Positive', 'Negative', 'Positive', 'Negative', 'Positive']})

# Define custom colors
custom_colors = ["#3498db", "#e74c3c"]

# Create a count plot of the 'Result' column
ax = sns.countplot(x='Result', data=df_copy, palette=custom_colors)  # Specify 'x' as the column name
plt.title('WHR-Distribution')

# Add labels to the bars
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), '{:.0f}'.format(p.get_height()), ha='center')

# Remove spines
sns.despine(left=True, bottom=True)

plt.show()
```

**The given count plot shows that in the World Happiness Report, the distribution of positive value is 3 whereas the distribution of negative value is 2**
